## Análisis de Nielsen

```python
df_nielsen.insert(0,'area de especialidad del evaluado',df_eval['En que área te especializas'])
```

```python
df_nielsen.head()
```

```python
df_nielsen_proc = df_nielsen.replace({"0. muy bueno": 0, "1. bueno": 1,'2. regular':2,'3. malo':3,'4. muy malo':4,'N/A':0})
```

```python
df_nielsen_data = df_nielsen_proc.drop(columns=['area de especialidad del evaluado'])
```

```python
nielsen_mean_cols = []
nielsen_mean_data = []
for col in df_nielsen_data.columns:
    nielsen_mean_cols.append(col[:-9])
    nielsen_mean_data.append(df_nielsen_data[col].mean().round(2))
```

```python
nielsen_mean = pd.Series(data=nielsen_mean_data, index=nielsen_mean_cols)
```

```python
sns.set(rc = {'figure.figsize':(8,8)})
ax = nielsen_mean.plot.barh()
for i in ax.containers:
    ax.bar_label(i,)
```

```python
df_nielsen_proc_2 = df_nielsen.drop(columns=['area de especialidad del evaluado'])
```

```python
df_nielsen_proc_2 =  df_nielsen_proc_2.replace({"0. muy bueno": '0. No es problema', "1. bueno": '1. Problemas apenas estéticos','2. regular':'2. Problemas menores de usabilidad','3. malo':'3. Problemas mayores de usabilidad','4. muy malo':'4. Catástrofe de usabilidad','N/A':'No aplica'})
```

```python
for col in df_nielsen_proc_2.columns:
    df_nielsen_proc_2.rename(columns={col:col[:-9]},inplace=True)
```

```python
arr_resultados = []
for col in df_nielsen_proc_2.columns:
    arr_resultados.append(df_nielsen_proc_2[col].value_counts().sort_index(ascending=True))
```

```python
df_res_nielsen = pd.DataFrame(arr_resultados)
```

```python
sns.set_style("dark")
sns.set(rc = {'figure.figsize':(15,5)})
ax = df_res_nielsen.plot(kind="bar")
for i in ax.containers:
    ax.bar_label(i,)
plt.title("Análisis de por heurística")
plt.xlabel("Heurísticas")
plt.ylabel("Puntuación")
plt.legend(bbox_to_anchor=(0.9, 1), loc='center', frameon=False,borderaxespad=0)
```

## Análisis de SUS

```python
df_sus['preguntas positivas'] = df_sus.iloc[:,0:9:2].sum(axis=1)-5
```

```python
df_sus['preguntas negativas'] = 25-df_sus.iloc[:,1:10:2].sum(axis=1)
```

```python
df_sus['Puntaje SUS'] = (df_sus['preguntas positivas']+df_sus['preguntas negativas'])*2.5
```

```python
df_sus.head()
```

```python
puntuacion_total = df_sus['Puntaje SUS'].mean()
```

```python
sus_mean_cols = []
sus_mean_data = []
for col in df_sus.columns:
    sus_mean_cols.append(col)
    sus_mean_data.append(df_sus[col].mean().round(2))
```

```python
sus_mean = pd.Series(data=sus_mean_data, index=sus_mean_cols)
```

```python
sus_mean = sus_mean[:-3]
```

```python
sus_mean
```

```python
arr_sus_resultados = []
for col in df_sus.columns[:-3]:
    arr_sus_resultados.append(df_sus[col].value_counts().sort_index(ascending=True))
df_res_sus = pd.DataFrame(arr_sus_resultados)
df_res_sus = df_res_sus.transpose().sort_index()
```

```python
new_cols =['1. Creo que usaría esta aplicación \nfrecuentemente',	
'2. Encuentro esta aplicación \ninnecesariamente complejo',	
'3. Creo que la aplicación fue \nfácil de usar',
'4. Creo que necesitaría ayuda de una \npersona con conocimientos técnicos \npara usar esta aplicación',	
'5. Encontré todas las funciones de la \naplicación bien integradas',	
'6. Esta aplicación contiene muchas \ninconsistencias en su contenido y \nfuncionalidad',	
'7. Imagino que la mayoría de la \ngente aprendería a usar esta \naplicación en forma muy rápida',	
'8. Encuentro que la aplicación es \nmuy difícil de usar',	
'9. Sentí confianza usando la aplicación',	
'10. Necesito aprender muchas \ncosas antes de usar esta aplicación']
```

```python
for i in range(len(df_res_sus.columns)):
    df_res_sus.rename(columns={df_res_sus.columns[i]:new_cols[i]},inplace=True)
```

```python
df_res_sus = df_res_sus.transpose()
```

```python
sns.set_style("dark")
sns.set(rc = {'figure.figsize':(14,4)})
ax = df_res_sus.plot(kind="bar")
for i in ax.containers:
    ax.bar_label(i,)
plt.title("Análisis de por heurística")
plt.xlabel("Heurísticas")
plt.ylabel("Puntuación")
plt.legend(bbox_to_anchor=(0.95, 1), loc='center', frameon=False,borderaxespad=0)
```
