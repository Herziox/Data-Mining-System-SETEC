# Describing Data

**Cantidad de datos.** Para la mayoría de las técnicas de modelado, existen compensaciones asociadas con el tamaño de los datos. Grandes conjuntos de datos pueden producir modelos más precisos, pero también pueden alargar el tiempo de procesamiento. Considere si usar un subconjunto de datos es una posibilidad. Cuando tome notas para el informe final, asegúrese de incluir estadísticas de tamaño para todos los conjuntos de datos y recuerde considerar tanto la cantidad de registros como los campos (atributos) al describir los datos.

```python
def amount_of_data(df):
    amount_of_data_list = []
    for col in df.columns:  
        amount_of_data_list.append(len(df[col][df[col].notnull()]))
    return amount_of_data_list

def null_of_data(df):
    null_of_data_list = []
    for col in df.columns:  
        null_of_data_list.append(len(df[col][df[col].isnull()]))
    return null_of_data_list

def not_null_of_dat(df):
    return df.notnull().sum().sum()

def n_row_data(df):
    return df.shape[0]

def n_col_data(df):
    return df.shape[1]
```

**Tipos de valor.** Los datos pueden tomar una variedad de formatos, como numérico, categórico (cadena) o booleano (verdadero/falso). Prestar atención al tipo de valor puede evitar problemas durante el modelado posterior.

```python
def value_types(df):
    value_types_list = df.dtypes.values
    return value_types_list
```

**Esquemas de codificación.** Con frecuencia, los valores de la base de datos son representaciones de características como el género o el tipo de producto. Por ejemplo, un conjunto de datos puede usar M y F para representar hombres y mujeres, mientras que otro puede usar los valores numéricos 1 y 2. Tenga en cuenta cualquier esquema en conflicto en el informe de datos.

```python
def coding_schemes(df):
    coding_schemes_list = df.iloc[0].values
    return coding_schemes_list
```

```python
info_df_coec = pd.DataFrame(data=[df_coec.columns,amount_of_data(df_coec),null_of_data(df_coec),value_types(df_coec),coding_schemes(df_coec)]).transpose()
info_df_coec = info_df_coec.rename(columns={0:'Nombre de columna',1:'# de datos NO nulos',2:'# de datos nulos',3:'Tipo de datos',4:'Ejemplo'})
```

```python
report_files_columns = ['Archivo','# Filas','# Columnas','Total datos', 'Total datos NO Nulos']
data_report = []
for path_file in os.listdir(path_wd_data):
    if '.csv' in path_file:
        df_check = pd.read_csv(os.path.join(path_wd_data,path_file))
        data_report_row = [path_file,n_row_data(df_check), n_col_data(df_check), n_row_data(df_check)*n_col_data(df_check), not_null_of_dat(df_check)]
        data_report.append(data_report_row)
        info_df = pd.DataFrame(data=[df_check.columns,amount_of_data(df_check),null_of_data(df_check),value_types(df_check),coding_schemes(df_check)]).transpose()
        info_df = info_df.rename(columns={0:'Nombre de columna',1:'# de datos NO nulos',2:'# de datos nulos',3:'Tipo de datos',4:'Ejemplo'})
        info_df.to_excel(os.path.join('data_understanding_reports','reporte_'+path_file[:-4]+'.xlsx'),index=False)

df_data_report = pd.DataFrame(data=data_report,columns=report_files_columns)
df_data_report.to_excel(os.path.join('data_understanding_reports','reporte_datos.xlsx'),index=False)
```
